```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from tqdm import tqdm
```

```python
df_improved = pd.read_csv("improved_gs_1.csv", index_col=0)
df_paper = pd.read_csv("paper_gs_1.csv", index_col=0)
```

```python
df_paper = df_paper.sort_values("val_top10_acc", ascending=False)[:10].reset_index(drop=True)
df_improved = df_improved.sort_values("val_top10_acc", ascending=False)[:10].reset_index(drop=True)
```

```python
df_paper["paper_val_top10_acc"] = df_paper["val_top10_acc"]
df_improved["improved_val_top10_acc"] = df_improved["val_top10_acc"]
```

```python
ax = df_paper["paper_val_top10_acc"].plot()
ax = df_improved["improved_val_top10_acc"].plot(ax=ax)
plt.legend()
```
